```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
```

```python
iris = sns.load_dataset("iris")
X = iris.drop('species',axis=1)
x=X['sepal_length']
```

```python
# A very nice trick to combine OOP plot with seaborn! 
from scipy.stats import kurtosis, skew


fig, axes = plt.subplots(2, sharex = True, gridspec_kw={"height_ratios": (0.2, 1)}, dpi=300)

sns.boxplot(x, ax= axes[0],color='yellow')
axes[0].axvline(x.mean(), color='r', linestyle='-.') 
axes[0].axvline(x.median(), color='g', linestyle='-.')
axes[0].axvline(x.mode()[0],  color='b', linestyle='-.') 
axes[0].set_title('Central tendencies of sepal_length \n Kurtosis= {a} \n Skewness = {b}'.format(a = np.round(kurtosis(x), decimals=2), b = np.round(skew(x), decimals=2) ))

sns.distplot(x, ax = axes[1])
axes[1].axvline(x.mean(), color='r', linestyle='-.') 
axes[1].axvline(x.median(), color='g', linestyle='-.')
axes[1].axvline(x.mode()[0],  color='b', linestyle='-.')  #x.mode() is a dataframe that is why sliced differently!



plt.legend({'Mean':x.mean(),'Median':x.median(),'Mode':x.mode()[0]})
```
